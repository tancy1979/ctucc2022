## Task 1. Magic Number

### Task 1.1 [3]
The following program converts a non-negative decimal number to its binary string equivalent. It currently does not cater for an edge case. Add program code with appropriate comments to cater for this edge case.

````python
# Denary/decimal to binary
def dec2bin(num):
  # declare result variable
  result = ''
  while num != 0: # loop while can still divide / quotient is non-zero 
    remainder = num % 2 # divide and get remainder
    num = num // 2 # reduce number / quotient by 2
    result = str(remainder) + result  # add remainder to front of result
  if result == '': # when num is 0
    result = '0'
  return result

# main
# print(dec2bin(23)) 
print(dec2bin(33)) # 0 is also a non-negative number
````

# Task 2. Sorting 

## Task 2.1 [6]

The following `bubble_sort()` function sorts items in ascending order. Provide appropriate comments in lines 4, 5, 7 - 10 to explain the purpose of the statements (not just literal translation of code).


````python
def bubble_sort(A):
  passes = len(A) - 1 # for n items, need n-1 passes
  swapped = True # assume not sorted
  while swapped: # enter loop if list is not sorted 
    swapped = False # assume not sorted first
    i = 1
    while i <= passes: # exit loop when i is bigger than passes
      if A[i-1] > A[i]: # condition to swap
        A[i-1], A[i] = A[i], A[i-1] # swap order of items, smaller in front
        swapped = True # if not true, then list is sorted, then exit loop
      i = i + 1
    passes = passes - 1
  return A

# main
A = [6, 5, 3, 1, 8, 7, 2, 4]
print(bubble_sort(A))
````

# Task 2.2 [4]

The following function `bubble_sort_noop()` is an non-optimised implementation of bubble sort by a novice programmer. Explain why the code is non-optimal and how it can be improved in the text cell that follows.


````python
# Bubble sort non-optimised

def bubble_sort_noop(A):
  for i in range(len(A)):
    for j in range(len(A)-1):
      if A[j] > A[j+1]: # not in order
        A[j], A[j+1] = A[j+1], A[j] # swap
  return A

# main
A = [6, 5, 3, 1, 8, 7, 2, 4]
print(bubble_sort_noop(A))
````

Your explanation: For each iteration of the main 'for' loop, there is a need to compare adjacent numbers in the list, starting from the first one each time. So the number of comparisons made is A(A-1).
